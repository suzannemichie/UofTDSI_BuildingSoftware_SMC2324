```python
import requests
import json
from pprint import pprint

# get secret from Colab
from google.colab import userdata
token = userdata.get('ghtoken')

# initialize request parameters
url = 'https://api.github.com/search/repositories?q=stars:>1&sort=stars&per_page=10'
headers = {'Authorization': 'Bearer ' + token}

# send request to GitHub
r = requests.get(url, headers=headers)

# print intial responses
print(r.status_code)
print(r.text)

r_json = json.loads(r.text)
pprint(r_json)
```

```python
items = r_json['items']
print(items[0])

repo_name = []
repo_stars = []
for item in items:
  repo_name.append(item['name'])
  repo_stars.append(item['stargazers_count'])

print(repo_name)
print(repo_stars)
```

```python
import matplotlib.pyplot as plt

plt.barh(repo_name, repo_stars)
```
